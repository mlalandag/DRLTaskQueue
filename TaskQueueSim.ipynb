```python
class TaskQueue():
    
    def __init__(self):
        self.number_of_tasks = 0
        self.queue           = collections.deque([]) 
        
    def add_task(task)
        self.number_of_tasks += 1
        self.queue.append(task)
        
    def take_task()
        if len(self.queue) > 0
            self.number_of_tasks -= 1
            return self.queue.popleft()
        
    def get_num_tasks():
        return self.number_of_tasks
    
```

0 - Add Worker to the pool
